```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

# 1. Масштабирование:
# а. Нормализовать признак - median_income
# б. Привести признак - population к масштабу от 0 до 100

HOUSING_PATH = "Материалы_к_лекции_1/data/housing.csv"
housing = pd.read_csv(HOUSING_PATH)
housing.head()
```

```python
#а. Нормализовать признак - median_income
min_max_scaler = preprocessing.MinMaxScaler()
housing_column_to_normalize = housing[["median_income"]]
housing_column_to_normalize.head()
housing_column_normalized = min_max_scaler.fit_transform(housing_column_to_normalize)
housing_column_normalized
```

```python
# б. Привести признак - population к масштабу от 0 до 100

min_max_scaler = preprocessing.MinMaxScaler(feature_range = (0,100))
housing_column_to_normalize = housing[["population"]]
housing_column_to_normalize.head()
housing_column_normalized = min_max_scaler.fit_transform(housing_column_to_normalize)
housing_column_normalized
```

```python
# 2. Заполнение пропусков:
# а. Заполнить пропуски в столбце total_bedrooms средними значениями по категориям столбца ocean_proximity
# Например, если пропущенно значение в строке с категорией INLAND, нужно заполнить его средним значением по этой категории


housing['total_bedrooms']= housing.groupby("ocean_proximity").transform(lambda x: x.fillna(x.mean()))

housing.head()
```

```python
# 3. Категорийные атрибуты:
# а. Посчитать среднее кол-во комнат в строке (total_rooms/households)
# б. Разбить полученный столбец на группы на ваше усмотрение (Например, 1 комната, 2 комнаты, 3 и более комнат)

housing['average_room_num'] = (housing.total_rooms/ housing.households).apply(round)
housing.head()
```

```python
# б. Разбить полученный столбец на группы на ваше усмотрение (Например, 1 комната, 2 комнаты, 3 и более комнат)

housing['rooms_in_house'] = pd.cut(housing['average_room_num'], bins=[0,1,2,99], labels=['1 команата', '2 комнаты','3 и более комнат'])

housing.sort_values("average_room_num")
```
